```python
from sage.matrix.berlekamp_massey import berlekamp_massey
import random
from matplotlib import pyplot as plt
```

```python
def random_vector(n, f):
    return vector(f, [f.random_element() for _ in range(n)])

def sparse_vector(k, n, f):
    # k is number of non-zero elements
    indices = random.sample(range(n), k)
    v = [f(0)] * n
    for i in indices:
        v[i] = f.random_element()
    return vector(f, v)

def structured_vector(n, f):
    return vector(f, [1]*n) 

def generate_vectors(category, n, f):
    if category == "Random":
        return random_vector(n, f), random_vector(n, f)
    elif category == "Sparse":
        return sparse_vector(1, n, f), sparse_vector(1, n, f)
    elif category == "Structured":
        return structured_vector(n, f), structured_vector(n, f)
    else:
        raise ValueError("Unknown category")
```

```python
def minpoly_recovery_success(A, f, trials=1000):
    n = A.nrows() # dimension of A
    success_rates = {"Random" : 0}
    min_poly = A.minpoly() # actual min poly
    for category in success_rates.keys():
        succeeded = 0
        for _ in range(trials):
            u, v = generate_vectors(category, n, f)
            kappa = [u.dot_product((A^i)*v) for i in range(2*n)
            mu = berlekamp_massey(kappa) 
            if mu == min_poly: 
                succeeded += 1
        success_rates[category] = f'{float(succeeded * 100/trials):.2f}%'
    return success_rates
```

### Experiment 1: Same Eigenvalue Multiplicities, Different Jordan Structures

**Parameters:**
- **Fixed:** Characteristic polynomial (x-1)²(x-2)(x-3)³, Field GF(7)
- **Varying:** Jordan block arrangements for each eigenvalue

**Jordan Block Configurations:**
- **Config A:** Single block per eigenvalue: [2] for λ=1, [1] for λ=2, [3] for λ=3
- **Config B:** Mixed blocks: [1,1] for λ=1, [1] for λ=2, [2,1] for λ=3  
- **Config C:** All minimal blocks: [1,1] for λ=1, [1] for λ=2, [1,1,1] for λ=3

```python
def multi_eigenvalue_jordan_matrix(eigenvalue_configs, f):
    F = f
    all_blocks = []
    
    for eigenvalue, block_sizes in eigenvalue_configs:
        for size in block_sizes:
            J = Matrix(F, size, size)
            for i in range(size):
                J[i, i] = eigenvalue
                if i + 1 < size:
                    J[i, i + 1] = 1
            all_blocks.append(J)
    
    return block_diagonal_matrix(all_blocks)
```

```python
test_data = {}


configurations = {
    "Config A: [2],[1],[3]": [(1, [2]), (2, [1]), (3, [3])],
    "Config B: [2],[1],[2,1]": [(1, [2]), (2, [1]), (3, [2,1])],
    "Config C: [2],[1],[1,2]": [(1, [2]), (2, [1]), (3, [1,2])],
    "Config D: [2],[1],[1,1,1]": [(1, [2]), (2, [1]), (3, [1,1,1])],
    "Config E: [1,1],[1],[3]": [(1, [1,1]), (2, [1]), (3, [3])],
    "Config F: [1,1],[1],[2,1]": [(1, [1,1]), (2, [1]), (3, [2,1])],
    "Config G: [1,1],[1],[1,2]": [(1, [1,1]), (2, [1]), (3, [1,2])],
    "Config H: [1,1],[1],[1,1,1]": [(1, [1,1]), (2, [1]), (3, [1,1,1])]
}

def test1(label, config):
    f = GF(7)
    A = multi_eigenvalue_jordan_matrix(config, f)
    success_rates = minpoly_recovery_success(A, f)
    test_data[label] = success_rates
    return success_rates

for label, config in configurations.items():
    test1(label, config)
    
plot("Experiment 1: Same Eigenvalue Multiplicities, Different Jordan Structures")
```

### Experiment 4: Jordan Block Permutation Effect

**Parameters:**
- **Fixed:** Characteristic polynomial (x-1)²(x-2)(x-3)³, Field GF(7), Jordan structure [2],[1],[2,1]
- **Varying:** Spatial ordering/permutation of Jordan blocks in the block diagonal matrix

**Research Question:** Does the ordering of Jordan blocks affect Krylov sequence generation and recovery success?

**Block Permutation Configurations:**
- **Canonical:** J₁([2]), J₂([1]), J₃([2]), J₃([1]) → grouped by eigenvalue
- **Interleaved:** J₁([2]), J₃([2]), J₂([1]), J₃([1]) → large blocks first
- **Reversed:** J₃([1]), J₃([2]), J₂([1]), J₁([2]) → reverse eigenvalue order
- **Mixed:** J₂([1]), J₁([2]), J₃([1]), J₃([2]) → alternating pattern

```python
def permuted_jordan_matrix(block_list, f):
    F = f
    blocks = []
    
    for eigenvalue, size in block_list:
        J = Matrix(F, size, size)
        for i in range(size):
            J[i, i] = eigenvalue
            if i + 1 < size:
                J[i, i + 1] = 1
        blocks.append(J)
    
    return block_diagonal_matrix(blocks)

test_data = {}

permutation_configs = {
    "Case 1": [(1, 2), (2, 1), (3, 2), (3, 1)],        # J₁([2]), J₂([1]), J₃([2]), J₃([1])
    "Case 2": [(1, 2), (3, 2), (2, 1), (3, 1)],      # J₁([2]), J₃([2]), J₂([1]), J₃([1])  
    "Case 3": [(3, 1), (3, 2), (2, 1), (1, 2)],         # J₃([1]), J₃([2]), J₂([1]), J₁([2])
    "Case 4": [(2, 1), (1, 2), (3, 1), (3, 2)]             # J₂([1]), J₁([2]), J₃([1]), J₃([2])
}

def test5(label, block_list):
    f = GF(7)
    A = permuted_jordan_matrix(block_list, f)
    success_rates = minpoly_recovery_success(A, f)
    test_data[label] = success_rates
    
    eigenvals = [ev for ev, size in block_list]
    block_sizes = [size for ev, size in block_list]
    print(f"{label}: blocks {list(zip(eigenvals, block_sizes))}")
    return success_rates

for label, block_list in permutation_configs.items():
    test5(label, block_list)
    
plot("Experiment 4: Jordan Block Permutation Effect")
```
