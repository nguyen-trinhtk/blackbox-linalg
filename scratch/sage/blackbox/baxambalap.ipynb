```python
def BM(q, sequence, n):
    f = GF(q)
    R = PolynomialRing(f, "x")
    sequence = [f(s) for s in sequence]
    C = [f(1)]
    B = [f(1)]
    L = 0
    m = 1
    streak = 1
    b = f(1)
    for k in range(len(sequence)):
        d = sequence[k] + sum(C[i] * sequence[k - i] for i in range(1, L + 1))
        if d == 0:
            m += 1
            streak += 1
        elif 2 * L <= k:
            T = C[:]
            C += [f(0)] * (m + len(B) - len(C))
            for i in range(len(B)):
                C[i + m] -= d / b * B[i]
            L = k + 1 - L
            B = T
            b = d
            m = 1
            streak = 1
        else:
            C += [f(0)] * (m + len(B) - len(C))
            for i in range(len(B)):
                C[i + m] -= d / b * B[i]
            m += 1
            streak = 1
    # result processing before return
    C = C[::-1]
    while len(C) > 1 and C[0] == 0:
        C = C[1:]
    return R(C)
```
